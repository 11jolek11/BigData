```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler, normalize, Normalizer

N = 500
t = np.arange(N)
x = np.random.normal(loc=1, scale=3, size=N)

plt.figure(figsize=(10,3))
plt.plot(t,x, label='x')
plt.legend()
plt.show()
```

# **Zad1**

```python
x = x.reshape(-1, 1)
```

```python
minmaxnormalizer = MinMaxScaler()
minmaxnormalized = minmaxnormalizer.fit_transform(x)

normalized = normalize(x, norm='l2')
# normalized = normalize(x, norm='l1')


plt.figure(figsize=(10,3))
plt.plot(t, normalized, label='normalized', color="tab:red")
plt.plot(t, x, label='orginal', color="tab:blue")

plt.legend()
plt.show()

print(np.max(normalized))
print(np.min(normalized))

plt.figure(figsize=(10,3))
plt.plot(t, minmaxnormalized, label='normalized', color="tab:red")
plt.plot(t, x, label='orginal', color="tab:blue")

plt.legend()
plt.show()
```

# Zad2

```python
scaler = StandardScaler()
scaler = scaler.fit(x)

stand = scaler.transform(x)
inversed = scaler.inverse_transform(normalized)

plt.figure(figsize=(10,3))
plt.plot(t, stand, label='stand', color="tab:red")
plt.plot(t, inversed, label='orginal', color="tab:blue")

plt.legend()
plt.show()

print(np.mean(stand))
print(np.std(stand))
```

```python
scaler = MinMaxScaler(feature_range=(-5, 5))
scaler = scaler.fit(x)

scaled = scaler.transform(x)

inversed = scaler.inverse_transform(scaled)


plt.figure(figsize=(10,3))
plt.plot(t, scaled, label='normalized', color="tab:red")
plt.plot(t, inversed, label='orginal', color="tab:blue")

plt.legend()
plt.show()

print(np.max(scaled))
print(np.min(scaled))
```
